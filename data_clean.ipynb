## Feature Engineering

#### Satisfaction Biner

```python
df['Satisfaction_Biner'] = df['Satisfaction'].apply(lambda x: 1 if x == 'Satisfied' else 0)
```

```python
df['Satisfaction_Biner'].head()
```

#### Grouping Age

```python
bins_age = [0, 25, 41, 61, df['Age'].max() + 1] 
labels_age = ['Youth', 'Young Adult', 'Mature Adult', 'Senior']

df['Age Group'] = pd.cut(df['Age'], bins=bins_age, labels=labels_age, right=False)
```

```python
df['Age Group'].head()
```

#### Grouping Distance

```python
bins_distance = [0, 500, 1500, df['Flight Distance'].max() + 1]
labels_distance = ['Short Haul', 'Medium Haul', 'Long Haul']

df['Distance Group'] = pd.cut(df['Flight Distance'], bins=bins_distance, labels=labels_distance, right=False)
```

```python
df['Distance Group'].head()
```

## Check Outlier

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Kolom yang akan diperiksa outlier-nya
kolom_delay = ['Departure Delay', 'Arrival Delay']

plt.figure(figsize=(10, 5))

# Membuat Box Plot
for i, col in enumerate(kolom_delay):
    plt.subplot(1, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Daftar kolom yang akan dibatasi (capping)
delay_cols = ['Departure Delay', 'Arrival Delay']

print("Nilai sebelum Capping:")
print(df[delay_cols].describe().loc[['max', 'mean']])

# --- Menerapkan Capping ---

for col in delay_cols:
    # 1. Hitung nilai Batas Atas (Persentil ke-99)
    # Persentil ke-99 adalah nilai di mana 99% data berada di bawahnya.
    upper_limit = df[col].quantile(0.99)
    
    # 2. Gantikan (Cap) nilai yang melebihi batas atas dengan nilai batas itu sendiri
    df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])
    
    print(f"\nKolom '{col}' dibatasi pada nilai: {upper_limit:.2f}")


print("\nNilai setelah Capping (Cek Mean dan Max):")
print(df[delay_cols].describe().loc[['max', 'mean']].round(2))
```

#### Analisis Segmentasi Kunci (Menjawab "Siapa?")

Tujuan: Menemukan kelompok mana yang paling puas atau tidak puas.

```python
# Fungsi untuk menghitung Persentase Puas per segmen
def get_satisfaction_percentage(df, segment_col):
    # Menghitung rata-rata Satisfaction_Biner (yang setara dengan Persentase Puas)
    return (df.groupby(segment_col)['Satisfaction_Biner'].mean() * 100).sort_values(ascending=False)

# Inisialisasi plot dengan tata letak 2 baris dan 2 kolom
plt.figure(figsize=(18, 12)) 

# --- A. Kepuasan vs. Tipe Perjalanan ---
plt.subplot(2, 2, 1) # Plot di posisi 1 (Baris 1, Kolom 1)
type_travel_sat = get_satisfaction_percentage(df, 'Type of Travel')
sns.barplot(x=type_travel_sat.index, y=type_travel_sat.values, palette='coolwarm')
plt.title('Kepuasan Berdasarkan Tipe Perjalanan', fontsize=14)
plt.ylabel('Persentase Puas (%)')
plt.ylim(0, 100)
plt.xlabel('') # Menghilangkan label x agar lebih bersih

# --- B. Kepuasan vs. Kelompok Usia ---
plt.subplot(2, 2, 2) # Plot di posisi 2 (Baris 1, Kolom 2)
age_group_sat = get_satisfaction_percentage(df, 'Age Group')
sns.barplot(x=age_group_sat.index, y=age_group_sat.values, palette='plasma')
plt.title('Kepuasan Berdasarkan Kelompok Usia', fontsize=14)
plt.ylabel('Persentase Puas (%)')
plt.ylim(0, 100)
plt.xlabel('') 

# --- C. Kepuasan vs. Kelas Penerbangan (Tambahan) ---
plt.subplot(2, 2, 3) # Plot di posisi 3 (Baris 2, Kolom 1)
class_sat = get_satisfaction_percentage(df, 'Class')
sns.barplot(x=class_sat.index, y=class_sat.values, palette='Greens_d')
plt.title('Kepuasan Berdasarkan Kelas Penerbangan', fontsize=14)
plt.ylabel('Persentase Puas (%)')
plt.ylim(0, 100)

# --- D. Kepuasan vs. Customer Type (Tambahan untuk melengkapi 4 segmen penting) ---
plt.subplot(2, 2, 4) # Plot di posisi 4 (Baris 2, Kolom 2)
customer_type_sat = get_satisfaction_percentage(df, 'Customer Type')
sns.barplot(x=customer_type_sat.index, y=customer_type_sat.values, palette='Reds_d')
plt.title('Kepuasan Berdasarkan Tipe Pelanggan', fontsize=14)
plt.ylabel('Persentase Puas (%)')
plt.ylim(0, 100)
plt.xticks(rotation=15)

plt.tight_layout() # Mengatur jarak antar plot agar tidak bertumpukan
plt.show()
```

ðŸŽ¯ Insight Kunci Segmentasi Kepuasan Penumpang
I. Segmentasi Berdasarkan Tipe Perjalanan
Business Travel adalah segmen yang paling puas (sekitar 70-75%), menunjukkan bahwa maskapai relatif berhasil dalam memenuhi ekspektasi pelancong dengan prioritas tinggi pada efisiensi dan kenyamanan premium.

Personal Travel adalah segmen yang paling rentan terhadap ketidakpuasan (tingkat kepuasan hanya sekitar 10-15%), mengindikasikan adanya kesenjangan serius pada layanan dasar atau sensitivitas harga yang tinggi di segmen ini.

#### Analisis Diagnostik: Korelasi & Gap Analysis (Mencari Root Causes)

A. Analisis Korelasi (Key Drivers)

```python
# Daftarkan kolom skor layanan
score_cols = [
    'Departure and Arrival Time Convenience', 'Ease of Online Booking', 
    'Check-in Service', 'Online Boarding', 'Gate Location', 
    'On-board Service', 'Seat Comfort', 'Leg Room Service', 
    'Cleanliness', 'Food and Drink', 'In-flight Service', 
    'In-flight Wifi Service', 'In-flight Entertainment', 'Baggage Handling'
]

# Hitung Matriks Korelasi (hanya kolom yang relevan)
analysis_cols = score_cols + ['Satisfaction_Biner']
correlation_matrix = df[analysis_cols].corr()

# Ambil korelasi hanya terhadap 'Satisfaction_Biner'
satisfaction_corr = correlation_matrix['Satisfaction_Biner'].sort_values(ascending=False).drop('Satisfaction_Biner')

# Visualisasi Heatmap Korelasi
plt.figure(figsize=(6, 8))
sns.heatmap(satisfaction_corr.to_frame(), annot=True, cmap='coolwarm', fmt=".2f", cbar=False)
plt.title('Korelasi Fitur Layanan dengan Kepuasan Penumpang')
plt.show()

print("\nTop 5 Key Drivers (Korelasi Terkuat):")
print(satisfaction_corr.head(5).round(3))
```

ðŸ”¬ Insight Kunci: Key Drivers Kepuasan (Korelasi)
Analisis ini menunjukkan faktor-faktor mana yang memiliki hubungan paling kuat dengan hasil kepuasan penumpang.

I. Prioritas Utama (High Impact Factors)
Fitur-fitur ini menunjukkan korelasi positif yang signifikan, menjadikannya Key Drivers yang paling efektif untuk memengaruhi hasil kepuasan penumpang:

Online Boarding (Ï=+0.50): Ini adalah faktor dengan korelasi terkuat. Hal ini mengindikasikan bahwa efisiensi proses digital sebelum penerbangan (kemudahan dan kecepatan online boarding) adalah penentu utama kepuasan total.

In-flight Entertainment (Ï=+0.40): Menunjukkan kontribusi signifikan, khususnya dalam mengelola persepsi durasi penerbangan. Peningkatan pada hiburan adalah lever yang kuat.

Seat Comfort (Ï=+0.35): Faktor kenyamanan fisik yang merupakan pengeluaran terbesar ketiga yang harus diatasi.

II. Faktor Pendukung Kuat
Faktor-faktor ini berkorelasi sedang dan penting sebagai layanan pendukung yang konsisten:

On-board Service (Ï=+0.32): Kualitas interaksi kru kabin secara langsung memengaruhi persepsi layanan.

Leg Room Service (Ï=+0.31): Kenyamanan ruang kaki adalah faktor fisik yang penting, terutama karena sering menjadi sumber keluhan di kelas ekonomi.

Cleanliness (Ï=+0.31): Kebersihan pesawat dan toilet memiliki korelasi yang substansial dengan kepuasan.

III. Faktor Non-Kritis / Non-Pengungkit
Faktor-faktor ini menunjukkan korelasi yang sangat lemah atau negatif, yang berarti perbaikan pada area ini kemungkinan tidak akan berdampak besar pada peningkatan kepuasan keseluruhan:

Gate Location (Ï=âˆ’0.00): Korelasi mendekati nol, menunjukkan bahwa lokasi gerbang tidak memiliki pengaruh yang signifikan terhadap persepsi kepuasan penumpang secara keseluruhan.

Departure and Arrival Time Convenience (Ï=âˆ’0.05): Korelasi negatif yang kecil mengindikasikan bahwa ketidaknyamanan waktu keberangkatan/kedatangan cenderung mendorong ketidakpuasan, tetapi convenience yang baik bukanlah pendorong utama kepuasan tinggi.

Ease of Online Booking (Ï=+0.17): Meskipun pemesanan mudah, ini bukan lagi faktor pembeda utama setelah penumpang naik ke pesawat.

B. Gap Analysis (Root Cause Finder)

```python
# 1. Hitung Rata-rata Skor per Kelompok (Puas vs. Tidak Puas)
# .T untuk transpose agar kolom layanan menjadi baris
avg_scores = df.groupby('Satisfaction_Biner')[score_cols].mean().T 

# 2. Hitung Gap Score
avg_scores.columns = ['Avg_Dissatisfied', 'Avg_Satisfied']
avg_scores['Gap_Score'] = avg_scores['Avg_Satisfied'] - avg_scores['Avg_Dissatisfied']

# 3. Ranking Root Causes
root_causes = avg_scores.sort_values(by='Gap_Score', ascending=False)

print("\n--- Ranking Root Causes (Top 5 Gap Score) ---")
print("Ini adalah fitur yang memiliki kesenjangan penilaian terbesar dan merupakan target perbaikan.")
print(root_causes[['Avg_Satisfied', 'Avg_Dissatisfied', 'Gap_Score']].head(5).round(2))

# Visualisasi Gap Score (Ini adalah visual utama untuk Dashboard PBI)
plt.figure(figsize=(10, 6))
sns.barplot(x=root_causes.index, y='Gap_Score', data=root_causes, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Ranking Root Causes (Gap Score Terbesar)')
plt.ylabel('Selisih Rata-rata Skor (Puas - Tidak Puas)')
plt.show()
```

Implikasi Kritis dari Gap Score
Prioritas Tertinggi: Konsistensi Digital: Fitur Online Boarding berada di peringkat teratas dengan Gap Score 1.37. Hal ini menunjukkan bahwa sistem online boarding berfungsi secara ekstrim tidak konsisten. Bagi penumpang yang tidak puas, skor Online Boarding sangat rendah (2.66), menunjukkan kegagalan sistem yang harus segera diperbaiki.

Faktor Nyaman Fisik: Tiga dari lima Root Cause teratas (Seat Comfort, On-board Service, Leg Room Service) adalah faktor kenyamanan fisik dan interaksi manusia. Perbaikan tidak hanya memerlukan investasi pada hardware tetapi juga pelatihan kru dan proses turnaround yang lebih baik.

III. Faktor yang Tidak Menjadi Akar Masalah (Low Impact)
Departure and Arrival Time Convenience (Gap Score Negatif: âˆ’0.20): Korelasi yang sangat lemah dan Gap Score negatif menunjukkan bahwa kenyamanan waktu kedatangan/keberangkatan bukanlah pendorong utama kepuasan atau ketidakpuasan, dibandingkan dengan kualitas layanan di pesawat.

Gate Location (Gap Score Rendah: +0.04): Kesenjangan skor yang mendekati nol memvalidasi temuan korelasi sebelumnya, bahwa lokasi gerbang bandara tidak memengaruhi pengalaman layanan maskapai secara signifikan.

Kesimpulan Utama (Root Cause):

Akar masalah utama yang harus menjadi fokus investasi dan kebijakan adalah inkonsistensi sistem digital (Online Boarding) dan kualitas pengalaman fisik di dalam pesawat (In-flight Entertainment dan Seat/Leg Room Comfort). Mengatasi kesenjangan pada fitur-fitur ini akan memberikan dampak terbesar dalam mengubah penumpang yang tidak puas menjadi puas.
